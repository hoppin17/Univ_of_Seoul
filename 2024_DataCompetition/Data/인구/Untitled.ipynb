```python
import pandas as pd
import numpy as np

def process_files(dir_list):
    all_data = []
    for i in range(1,11):
        try:
            df = pd.read_csv(dir_list[i], encoding='cp949')
            second_row = df.iloc[2]  # 2번째 행의 데이터 가져오기 (인덱스 1)
            all_data.extend(second_row.values)
        except Exception as e:
            print(f"Error reading file {dir_list[i]}: {e}")
    
    # 27개 단위로 끊어서 새로운 DataFrame 생성
    chunks = [all_data[i:i+27] for i in range(0, len(all_data), 27)]
    return pd.DataFrame(chunks)

# 10개의 파일 세트에 대해 처리
final_df = pd.DataFrame()
# for i in range(10):
#     dir_list = [f'set{i+1}_file{j+1}.csv' for j in range(10)]  # 각 세트의 파일 경로
#     df = process_files(dir_list)
#     final_df = pd.concat([final_df, df], ignore_index=True)

df = process_files(dir_list)

# 결과 확인
print(final_df.shape)
print(final_df.head())
```

```python
# CSV 파일 읽기
df = pd.read_csv('../aux_data_MonthlyCar.csv')

# 년도와 월을 기준으로 그룹화
df['year_month'] = df['year'].astype(str) + '_' + df['month'].astype(str)

# 각 년월 그룹 내에서 NA 값을 최소값으로 대체
columns_to_fill = ['car_sedan', 'car_van', 'car_twowheels', 'car_special', 'car_truck']

for col in columns_to_fill:
    df[col] = df.groupby('year_month')[col].transform(lambda x: x.fillna(x.min()))

df

# 결과를 CSV 파일로 저장 (선택사항)
df.to_csv('../updated_aux_data_MonthlyCar.csv', index=False)
```

```python
# 데이터 로드
df = pd.read_csv('../aux_data_DailyRain.csv')

# 날짜 열을 datetime 형식으로 변환
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')

# 월별, 지역별로 그룹화하고 강우 횟수와 총 강우량 계산
result = df.groupby([df['date'].dt.to_period('M'), 'gu']).agg({
    'rain_day': [('rain_count', lambda x: (x > 0).sum()),
                 ('total_rain', 'sum')]
}).reset_index()

# 결과 출력
print(result)
```
