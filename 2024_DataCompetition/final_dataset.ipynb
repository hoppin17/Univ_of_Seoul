```python
min_values = train_temp.groupby(['year', 'month', 'gu'])['tax'].min().reset_index()
min_values.columns = ['year', 'month', 'gu', 'min_value']

# 최소값의 절반 계산
min_values['half_min'] = min_values['min_value'] / 2

# train_temp의 각 그룹별로 NaN 값을 최소값의 절반으로 채우기
for index, row in min_values.iterrows():
    mask = (train_temp['year'] == row['year']) & \
           (train_temp['month'] == row['month']) & \
           (train_temp['gu'] == row['gu']) & \
           (train_temp['tax'].isnull())
    train_temp.loc[mask, 'tax'] = row['half_min']

# 결과 출력
print(train_temp)
```

```python
min_values = test_temp.groupby(['year', 'month', 'gu'])['tax'].min().reset_index()
min_values.columns = ['year', 'month', 'gu', 'min_value']

# 최소값의 절반 계산
min_values['half_min'] = min_values['min_value'] / 2

# train_temp의 각 그룹별로 NaN 값을 최소값의 절반으로 채우기
for index, row in min_values.iterrows():
    mask = (train_temp['year'] == row['year']) & \
           (train_temp['month'] == row['month']) & \
           (train_temp['gu'] == row['gu']) & \
           (train_temp['tax'].isnull())
    test_temp.loc[mask, 'tax'] = row['half_min']

# 결과 출력
print(test_temp)
```
